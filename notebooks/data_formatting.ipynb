```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("./imessages_high_sierra.csv")  
```

```python
df.head()
```

```python
data = df[["text", "is_sent"]][1:].to_numpy()
```

```python
i = 0
questions = []
answers = []
delim = " <|brk|> "
while (i < len(data)):
    q = ""
    a = ""
    
    while (i < len(data)):
        text = data[i][0]
        is_sent = data[i][1]
        if is_sent == 1:
            break;
        q += text + delim
        i += 1
    q = q[:-len(delim)]
    
    while (i < len(data)):
        text = data[i][0]
        is_sent = data[i][1]
        if is_sent == 0:
            break;
        a += text + delim
        i += 1
    a = a[:-len(delim)]
    
    questions.append(q)
    answers.append(a)
# print(questions[:2])
```

```python
df_new = pd.DataFrame({"questions": questions, "answers": answers})
```

```python
df_new.to_csv('./questions_answers.csv', index = False, encoding='utf-8')
```
